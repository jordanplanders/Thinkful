1. If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 or 5 below 1000.

```python
looking = 1
ik = 1000
while looking< 14:
    if (ik %3 == 0):
        three_val = ik
        looking = looking *3
    elif ik %5 == 0:
        five_val = ik
        looking = looking *5
    ik -=1

im = 0
for ip in range(int(three_val/3)):
    im += ip*3
    
for ip in range(int(five_val/5)):
    im += ip*5
    
print(im)
```

2. Each new term in the Fibonacci sequence is generated by adding the previous two terms. By considering the terms in the Fibonacci sequence whose values do not exceed four million, find the sum of the even-valued terms.

```python
n1, n2 = 1,2
fibs = [n2]

def fib(n1, n2):
    if n2+n1>4*10**6:
        return 'Done'
    else:
        if (n1+n2) %2 == 0:
            fibs.append(n1+n2)
        return fib(n2, n1+n2)

fib(n1, n2)
sum(fibs)
```

4. A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two 3-digit numbers.

```python
def find_palindrome():
    for ik in range(999,100,-1):
        for im in range(999,100,-1):
            num_str = list(str(ik*im))
            if [num_str[0],num_str[1]] == [num_str[-2],num_str[-1]]:
                return ik*im, ik, im

find_palindrome()
```

9. A Pythagorean triplet is a set of three natural numbers, a < b < c, for which, a2 + b2 = c2. There exists exactly one Pythagorean triplet for which a + b + c = 1000. Find the product abc.

```python
def find_triple():
    for a in range(1,500):
        for b in range(a,500):
            c = float((a**2+b**2)**(1/2))
            if c%1 == 0.0:
                if a+b+c == 1000:
                    return [a*b*c]
```

10. The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below two million.

```python
def find_prime(ik):
    im = 2
    while im<ik-1:
        if ik % im == 0:
            return [0]
        else:
            im+=1
    return [ik]


prime_sum = 5
value = 4
while value<100000:#2*10**6:
    prime = find_prime(value)
    prime_sum += prime[0]
    value +=1

prime_sum
```

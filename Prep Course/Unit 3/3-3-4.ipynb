```python
d_dist = {'normal': np.random.normal(5, 10 , n), 'poisson': np.random.poisson(2, n), 
              'bernoulli': np.random.binomial(1, .5, n),'binomial': np.random.binomial(20, 0.5, n), 
              'gamma': np.random.gamma(5,1, n), 'logistic': np.random.logistic(10, 1, n)}

dist_names = ['normal', 'poisson', 'bernoulli', 'binomial', 'gamma', 'logistic']
```

```python
def dist_breakdown(rand_dist, dist_name, n):
    # produce histogram
    plt.hist(rand_dist, bins=20, color='c', align='left')
    # compute mean and sigma & plot
    print('mean = ', rand_dist.mean(), ';', 'standard deviation = ', rand_dist.std())
    plt.axvline(rand_dist.mean(), color='b', linestyle='solid', linewidth=2)
    plt.axvline(rand_dist.mean() + rand_dist.std(), color='b', linestyle='dashed', linewidth=2)
    plt.axvline(rand_dist.mean()-rand_dist.std(), color='b', linestyle='dashed', linewidth=2) 
    plt.title(dist_name)
    plt.show()
```

```python
# generate random variable with 100 datapoints
for dist_name in dist_names:
    dist_breakdown(d_dist[dist_name],dist_name, 10000)
```

At n=100 it is difficult to see what the true behavior of the distribution is, but by increasing n to, say, 10000, it became more apparent (by inspection) that the mean and standard deviation are only meaningful when the distribution is continuous and approximates a normal distribution.

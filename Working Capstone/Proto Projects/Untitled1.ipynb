```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
th = np.linspace(0, 2*np.pi, 128)

fig, ax = plt.subplots(figsize=(3, 3))

# ax.set_title('style: {!r}'.form, color='C0')

ax.plot(th, np.cos(th), 'C1', label='C1', color = '#e50000')
ax.plot(th, np.sin(th), 'C2', label='C2', color = '#3f008e')
ax.legend()
```

```python
import matplotlib.pyplot as plt
cm = plt.cm.get_cmap('brg')
# xy = range(10)
x = [3,4,5,2,1,4,8,9,6,6,7,2,1]
y = [3,3,7,2,1,7,8,9,5,6,6,3,1]
z = [x[ik]+y[ik] for ik in range(len(x))]
sc = plt.scatter(x, y, c=z, vmin=0, vmax=2*max(z), s=35, cmap=cm)
plt.colorbar(sc)
plt.show()
```
